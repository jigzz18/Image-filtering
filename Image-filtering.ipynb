```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import skimage
import skimage.color as skic
import skimage.filters as skif
import skimage.data as skid
import skimage.util as sku
%matplotlib inline
```

```python
def show(img):
    fig, ax = plt.subplots(1, 1, figsize=(8, 8))
    ax.imshow(img, cmap=plt.cm.gray)
    ax.set_axis_off()
    plt.show()
```

```python
trudeau_img = mpimg.imread("trudeau.jpg")

img = skic.rgb2gray(trudeau_img)
show(img)
```

```python
show(skif.gaussian(img, 5.))
```

```python
img = skimage.img_as_float(mpimg.imread("trudeau.jpg"))


# We add Gaussian noise.
img_n = sku.random_noise(img)
show(img_n)

img_m= skic.rgb2gray(sku.random_noise(img))
show(img_m)
```

```python
img_r = skimage.restoration.denoise_tv_bregman(
    img_n, 5.)

fig, (ax1, ax2, ax3) = plt.subplots(
    1, 3, figsize=(12, 8))

ax1.imshow(img_n)
ax1.set_title('With noise')
ax1.set_axis_off()

ax2.imshow(img_r)
ax2.set_title('Denoised')
ax2.set_axis_off()

ax3.imshow(img)
ax3.set_title('Original')
ax3.set_axis_off()
```
